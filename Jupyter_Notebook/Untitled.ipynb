```python
from skimage import io
```

```python
photo = io.imread('VTNWZc.jpg')
print(type(photo))
```

```python
photo.shape
```

```python
from matplotlib import pyplot as plt
plt.imshow(photo)
plt.show()
```

```python
vflipped_photo = photo[:,::-1] # lateral inversion(vertical flip)
plt.imshow(vflipped_photo)
```

```python
hflipped_photo = photo[::-1,:] # horizontal inversion(horizontal flip)
plt.imshow(hflipped_photo)
```

```python
cropped_photo = photo[:500,:800] #Crop the image from photo
plt.imshow(cropped_photo)
```

```python
lower_cropped_photo = photo[:1600,1100:1800,::-1] #Crop the image from photo
plt.imshow(lower_cropped_photo)
```
