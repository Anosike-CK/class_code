```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('films.csv')
df = data[0:10][['revenue','vote_average']]
# vote_avg = data[0:10]['vote_average']
df
```

```python
revenue = df.revenue
vote_average = df.vote_average

plt.xlabel('vote average')
plt.ylabel('revenue')

plt.plot(vote_average,revenue)
```

```python
plt.bar(vote_average, revenue)
```

```python
plt.xlabel('vote average')
plt.ylabel('Revenue (in millions)')
plt.title('Movie revenue to vote distribution')

plt.bar(vote_average, revenue/1000000, width = 0.1, color = ['red'])
```

```python
plt.xlabel('vote average')
plt.ylabel('Revenue (in millions)')
plt.title('Movie revenue to vote distribution')

plt.barh(vote_average, revenue/1000000, color = ['red'])
```
